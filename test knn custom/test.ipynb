```python
import sklearn as sk
import pandas as pd
import numpy as np
```

```python
from sklearn import datasets

data=datasets.load_digits()
print(data.DESCR)
```

```python
X = data.data
# store the target data
y = data.target
# split the data using Scikit-Learn's train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,stratify=y,test_size=0.33,shuffle=True)
```

```python
print("media: ",X_train.mean())
print("deviazione std: ",X_test.std())
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
print("media: ",X_train.mean())
print("deviazione std: ",X_test.std())
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.utils import shuffle



logreg = KNeighborsClassifier(n_neighbors=3,n_jobs=-1)
logreg.fit(X_train, y_train)
pred=logreg.predict(X_test)
print(accuracy_score(y_test,pred))
from sklearn.metrics import classification_report
report =classification_report(y_test,pred)
print(report)
```

```python
def distanzaEuclidea(y,x):
  return np.linalg.norm(y-x)

def predizione_KNN_classification(X_training,X_test,Y_training,k):
  predizione=[]
  for x in X_test:
    pred,distanza,indici=KNN_classification(X_training,x,Y_training,k)
    predizione.append(pred)
  return predizione

def KNN_classification(X_training,X_test,Y_training,k):
  distanza=[]
  for x in X_training:
    distanza.append(distanzaEuclidea(X_test,x))
  indici=np.argsort(distanza)[:k]
  k_nearest=[]
  for i in indici:
    k_nearest.append(Y_training[i])
  valore_frequente=max(set(k_nearest), key=k_nearest.count)
  return valore_frequente,distanza,indici
```

```python
pred=predizione_KNN_classification(X_train,X_test,y_train,k=3)
print("custom: ",accuracy_score(y_test,pred))
report =classification_report(y_test,pred)
print(report)
```
